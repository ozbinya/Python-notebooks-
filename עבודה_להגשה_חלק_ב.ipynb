```python
# prompt: apply get dummy on the colmn Location

df = pd.get_dummies(df, columns=['Location'],dtype= int)
df.head()
```

```python
# prompt: map the Parking Availability coulmn to 0 and 1

# Assuming 'Parking Availability' column has string values like 'Yes' and 'No'
df['Parking Availability'] = df['Parking Availability'].map({'Yes': 1, 'No': 0})
df.head()
```

```python
df.describe()
```

```python
x= df.drop('Revenue',axis=1)
y= df['Revenue']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

הנתונים דורשים נרמול כיוון שהם מציגים נתונים בפרופורציות שונות.

באלגוריתמים knn ורגרסיה ליניראית קריטי לנרמל את הנתונים כיוון שהם מספריים ויכולים להיות בפרופורציות שונות אך במודל עץ החלטה המשתנים הם קטגוריאליים ולכן פחות קריטי לנרמל.


```python
x_scaler = MinMaxScaler()
x_train_scaled = x_scaler.fit_transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

```python
# prompt: get from numpy to dataframe

import pandas as pd
import numpy as np

# Assuming 'x_train_scaled' is your NumPy array
# and you want to convert it to a Pandas DataFrame

x_train_df = pd.DataFrame(x_train_scaled, columns=x_train.columns)
```

```python
# prompt: run linear regression model on the df include
# Root Mean Square Error (RMSE)
# Mean Absolut Error (MAE)
# R square
# the train and the test
# Mean Absolut Percent Error (MAPE)

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(x_train_scaled, y_train)

# Make predictions on the training and testing sets
y_pred_train = model.predict(x_train_scaled)
y_pred_test = model.predict(x_test_scaled)

# Calculate evaluation metrics
rmse_train = root_mean_squared_error(y_train, y_pred_train)
mae_train = mean_absolute_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)
mape_train = np.mean(np.abs((y_train - y_pred_train) / y_train)) * 100

rmse_test = root_mean_squared_error(y_test, y_pred_test)
mae_test = mean_absolute_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)
mape_test = np.mean(np.abs((y_test - y_pred_test) / y_test)) * 100


print("Train Set:")
print(f"RMSE: {rmse_train}")
print(f"MAE: {mae_train}")
print(f"R-squared: {r2_train}")
print(f"MAPE: {mape_train}")

print("\nTest Set:")
print(f"RMSE: {rmse_test}")
print(f"MAE: {mae_test}")
print(f"R-squared: {r2_test}")
print(f"MAPE: {mape_test}")
```

```python
# prompt: run polynomial regression on the train and the test

# Polynomial Regression
poly = PolynomialFeatures(degree=2) # Example: degree 2 polynomial features
x_train_poly = poly.fit_transform(x_train_scaled)
x_test_poly = poly.transform(x_test_scaled)

model_poly = LinearRegression()
model_poly.fit(x_train_poly, y_train)

y_pred_train_poly = model_poly.predict(x_train_poly)
y_pred_test_poly = model_poly.predict(x_test_poly)

# Evaluate polynomial regression model
rmse_train_poly = root_mean_squared_error(y_train, y_pred_train_poly)
mae_train_poly = mean_absolute_error(y_train, y_pred_train_poly)
r2_train_poly = r2_score(y_train, y_pred_train_poly)
mape_train_poly = np.mean(np.abs((y_train - y_pred_train_poly) / y_train)) * 100

rmse_test_poly = root_mean_squared_error(y_test, y_pred_test_poly)
mae_test_poly = mean_absolute_error(y_test, y_pred_test_poly)
r2_test_poly = r2_score(y_test, y_pred_test_poly)
mape_test_poly = np.mean(np.abs((y_test - y_pred_test_poly) / y_test)) * 100

print("\nPolynomial Regression - Train Set:")
print(f"RMSE: {rmse_train_poly}")
print(f"MAE: {mae_train_poly}")
print(f"R-squared: {r2_train_poly}")
print(f"MAPE: {mape_train_poly}")

print("\nPolynomial Regression - Test Set:")
print(f"RMSE: {rmse_test_poly}")
print(f"MAE: {mae_test_poly}")
print(f"R-squared: {r2_test_poly}")
print(f"MAPE: {mape_test_poly}")
```

```python
# prompt: run knn regression on the train and the test

# KNN Regression
knn_model = KNeighborsRegressor(n_neighbors=5)  # You can adjust the number of neighbors
knn_model.fit(x_train_scaled, y_train)

y_pred_train_knn = knn_model.predict(x_train_scaled)
y_pred_test_knn = knn_model.predict(x_test_scaled)

# Evaluate KNN regression model
rmse_train_knn = root_mean_squared_error(y_train, y_pred_train_knn)
mae_train_knn = mean_absolute_error(y_train, y_pred_train_knn)
r2_train_knn = r2_score(y_train, y_pred_train_knn)
mape_train_knn = np.mean(np.abs((y_train - y_pred_train_knn) / y_train)) * 100

rmse_test_knn = root_mean_squared_error(y_test, y_pred_test_knn)
mae_test_knn = mean_absolute_error(y_test, y_pred_test_knn)
r2_test_knn = r2_score(y_test, y_pred_test_knn)
mape_test_knn = np.mean(np.abs((y_test - y_pred_test_knn) / y_test)) * 100

print("\nKNN Regression - Train Set:")
print(f"RMSE: {rmse_train_knn}")
print(f"MAE: {mae_train_knn}")
print(f"R-squared: {r2_train_knn}")
print(f"MAPE: {mape_train_knn}")

print("\nKNN Regression - Test Set:")
print(f"RMSE: {rmse_test_knn}")
print(f"MAE: {mae_test_knn}")
print(f"R-squared: {r2_test_knn}")
print(f"MAPE: {mape_test_knn}")
```

```python
# prompt: run lasso regression on this data

from sklearn.linear_model import Lasso

# Initialize and train the Lasso Regression model
lasso_model = Lasso(alpha=1.0)  # You can adjust the alpha (regularization strength)
lasso_model.fit(x_train_scaled, y_train)

# Make predictions
y_pred_train_lasso = lasso_model.predict(x_train_scaled)
y_pred_test_lasso = lasso_model.predict(x_test_scaled)

# Evaluate Lasso Regression
rmse_train_lasso = root_mean_squared_error(y_train, y_pred_train_lasso)
mae_train_lasso = mean_absolute_error(y_train, y_pred_train_lasso)
r2_train_lasso = r2_score(y_train, y_pred_train_lasso)
mape_train_lasso = np.mean(np.abs((y_train - y_pred_train_lasso) / y_train)) * 100

rmse_test_lasso = root_mean_squared_error(y_test, y_pred_test_lasso)
mae_test_lasso = mean_absolute_error(y_test, y_pred_test_lasso)
r2_test_lasso = r2_score(y_test, y_pred_test_lasso)
mape_test_lasso = np.mean(np.abs((y_test - y_pred_test_lasso) / y_test)) * 100

print("\nLasso Regression - Train Set:")
print(f"RMSE: {rmse_train_lasso}")
print(f"MAE: {mae_train_lasso}")
print(f"R-squared: {r2_train_lasso}")
print(f"MAPE: {mape_train_lasso}")

print("\nLasso Regression - Test Set:")
print(f"RMSE: {rmse_test_lasso}")
print(f"MAE: {mae_test_lasso}")
print(f"R-squared: {r2_test_lasso}")
print(f"MAPE: {mape_test_lasso}")
```
